```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df=pd.read_csv('tmdb_movies.csv')
df1=pd.read_csv('tmdb_genres.csv')


#PODPUNKT1

table1=df.drop(columns=['Unnamed: 0','budget','homepage','id','original_language','overview','popularity','release_date','revenue','runtime','status','tagline'],inplace=False,axis=1)
table1.describe()

#Kwartyl trzeci wynosi 737.

table1.drop(table1[table1['vote_count'] <= 737].index, inplace = True)
table1.sort_values(by=['vote_average'],ascending=False,inplace=True)
table1.reset_index(inplace=True)
table1.drop(columns=['index'],axis=1,inplace=True)
table1.drop(labels=range(10,1200),axis=0,inplace=True)


#PODPUNKT2

table2=df.drop(columns=['Unnamed: 0', 'homepage', 'id','original_language','overview', 'popularity','runtime','tagline','vote_average','vote_count','genre_id'], inplace=False,axis=1)
table2.sort_values(by='release_date',ascending=True,inplace=True)
table2['release_date'] = pd.to_datetime(table2['release_date'])
table2.drop(table2[table2['release_date']<'2010-01-01'].index, inplace=True)
table2.drop(table2[table2['release_date']>'2016-12-31'].index, inplace=True)
table2.reset_index(inplace=True)
table2.drop(columns=['index'],axis=1, inplace=True)
table3=table2.groupby(pd.Grouper(key='release_date',freq='Y')).agg({'budget':'sum','original_title':'count'})

def average_count_for_budget(row):
    
    average=row['budget']/row['original_title']
    return average

table3['average_budget']=table3.apply(lambda row: average_count_for_budget(row),axis=1)


table4=table2.groupby(pd.Grouper(key='release_date',freq='Y')).agg({'revenue':'sum','original_title':'count'})

def average_count_for_revenue(row):
    
    average1=row['revenue']/row['original_title']
    return average1

table4['average_revenue']=table4.apply(lambda row: average_count_for_revenue(row),axis=1)
table3.drop(columns=['budget','original_title'],inplace=True,axis=1)
table4.drop(columns=['revenue','original_title'],inplace=True,axis=1)

table5=pd.concat([table3,table4],axis=1)
table5['year']=['2010','2011','2012','2013','2014','2015','2016']


#tworzenie wykresu

x=table5['year']
y=table5['average_budget']

x1=table5['year']
y1=table5['average_revenue']

def million(x, pos):
        return 'PLN {:2.1f}M'.format(x*1e-6)

fig, ax = plt.subplots()

formatter = plt.FuncFormatter(million)

ax.yaxis.set_major_formatter(formatter)
ax.bar(x1, y1,label='revenue')

ax.plot(x,y,color='r',label='budget')
ax.legend(loc=(1.05,0.85))
plt.title('Średni przychód i budżet filmu w latach 2010-2016')


#PODPUNKT3

table6=df.drop(columns=['Unnamed: 0', 'title', 'budget','revenue','status','release_date', 'homepage', 'id','original_language','overview', 'popularity', 'tagline', 'vote_average','vote_count'],inplace=False,axis=1)
table6.dropna(inplace=True)
table6.reset_index(inplace=True)
table6.drop(columns=['index'],axis=1,inplace=True)
table6=table6.astype({'genre_id': int})

df1.rename(columns={'Unnamed: 0':'genre_id'},inplace=True)
df1.dropna(inplace=True)
df1=df1.astype({'genre_id': int})

table7=pd.merge(table6,df1,how='inner',on='genre_id')


#PODPUNKT4

table8=table7.groupby(pd.Grouper(key='genres')).agg({'original_title':'count'})
table8.sort_values(by='original_title',ascending=False,inplace=True)
table8

#PODPUNKT5

table9=table7.groupby(pd.Grouper(key='genres')).agg({'runtime':'sum','original_title':'count'})

def average_runtime(row):
    
    average2=row['runtime']/row['original_title']
    return average2

table9['average_runtime']=table9.apply(lambda row: average_runtime(row),axis=1)

table9.sort_values(by='average_runtime',ascending=False,inplace=True)

#Najdłużej trwają filmy z kategorii "History"

#PODPUNKT6

table10=table7[table7['genres']=='History']
table10['runtime'].plot(kind='hist',bins=30)









```
